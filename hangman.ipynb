```python
import random
import time

f = open('data.txt')
bank = []
for line in f:
  line = line.replace('\n', '')
  if len(line) != 0:
    bank.append(line.lower())

def create_secret():
    return random.choice(bank)


SECRET = create_secret()
n = len(SECRET)

FIELD = [' ',
r'''





_______/|\_
''',
r'''

        |
        |
        |
        |
_______/|\_
''',
r'''
   +----+
        |
        |
        |
        |
_______/|\_
''',
r'''
   +----+
   |    |
   o    |
        |
        |
_______/|\_
''',
r'''
   +----+
   |    |
   o    |
  /|\   |
        |
_______/|\_
''',
r'''
   +----+
   |    |
   o    |
  /|\   |
  / \   |
_______/|\_
''']

# здесь мы наверное хотим иметь исходное поле
# и понимание, как оно меняется после каждого хода
mistakes = 0

right = []
mis = []
while mistakes < len(FIELD)-1:
    # make a move!
    print(FIELD[mistakes])
    for i in SECRET:
        if i in right:
            print(i, end = '')
        else:
            print('_', end = '')
    print()
    print(' '.join(mis))
    letter = input('Введите вашу букву: ')
    if letter not in SECRET:
        mistakes += 1
        mis.append(letter)
        mis = list(set(mis))
    else:
        right.append(letter)
        right = list(set(right))
    time.sleep(1)
if mistakes > len(FIELD)-2:
  print("Вы проиграли :(")
else:
  print("Вы выиграли!")
```
